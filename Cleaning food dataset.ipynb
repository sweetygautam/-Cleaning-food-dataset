```python
df_obj['GPA']=df_obj['GPA'].str.replace('bitch'," ")
```

```python
df_obj['GPA'].replace({'Unknown':'NaN','Personal':'NaN'},inplace=True)
```

```python
df_obj['GPA']=df_obj['GPA'].str.replace('Unknown','NaN').str.replace('Personal','NaN')
```

```python
df_obj.GPA.value_counts()
```

```python
df_obj['GPA'].mode()
```

```python
df_obj['GPA']=df_obj['GPA'].str.replace('NaN','3.5')
```

```python
df_obj.GPA.isna().sum()
```

```python
df_obj['GPA'].unique()
```

```python
df_obj.isna().sum()
```

```python
for i in df_obj:
    if df_obj[i].isna().sum()>0:
        df_obj[i].fillna(df_obj[i].mode()[0],inplace=True)
```

```python
df_obj.isna().sum()
```

```python
df_float.isna().sum()
```

```python
for i in df_float:
    if df_float[i].isna().sum()>0:
        df_float[i].fillna(df_float[i].mode()[0],inplace=True)
```
